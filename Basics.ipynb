# Star Wars Analysis

The data has several columns, including:

    RespondentID - An anonymized ID for the respondent (person taking the survey)
    Gender - The respondent's gender
    Age - The respondent's age
    Household Income - The respondent's income
    Education - The respondent's education level
    Location (Census Region) - The respondent's location
    Have you seen any of the 6 films in the Star Wars franchise? - Has a Yes or No response
    Do you consider yourself to be a fan of the Star Wars film franchise? - Has a Yes or No response


```python
import pandas as pd
import numpy as np

star_wars = pd.read_csv("star_wars.csv", encoding = "ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head(10)
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map({"Yes" : True, "No" : False})
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map({"Yes" : True, "No" : False})
```

```python
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "Star Wars: Episode I The Phantom Menace",
        "Unnamed: 10": "Star Wars: Episode II Attack of the Clones",
        "Unnamed: 11": "Star Wars: Episode III Revenge of the Sith",
        "Unnamed: 12": "Star Wars: Episode IV A New Hope",
        "Unnamed: 13": "Star Wars: Episode V The Empire Strikes Back",
        "Unnamed: 14": "Star Wars: Episode VI Return of the Jedi"
        })
```

```python
star_wars.head()
```

```python
cols = ["Star Wars: Episode I The Phantom Menace",
        "Star Wars: Episode II Attack of the Clones",
        "Star Wars: Episode III Revenge of the Sith",
        "Star Wars: Episode IV A New Hope",
        "Star Wars: Episode V The Empire Strikes Back",
        "Star Wars: Episode VI Return of the Jedi"]
means = star_wars[cols].mean()
means
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(range(6), means)
```

The lower the value, the better the rating. In this case, Empire Strikes Back seems to have been rated as the best
